```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
grades = pd.read_csv("GCC113_P2.csv", index_col=0)
```

```python
grades.head()
```

```python
# Students above average (60%)
label = ['< 60%', '>= 60%']
num_above = len(grades[grades.Total > 60])
num_below = len(grades) - num_above
plt.figure(figsize=((9,7)))
plt.bar(label, [num_below, num_above], color=['g', 'y'])
plt.ylabel('Number of grades')
plt.yticks(np.arange(20))
plt.grid(axis='y')
```

```python
# Distribution of the grades
plt.figure(figsize=((10,5)))
n, bins, patches = plt.hist(x=grades.Total, bins=10, color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.xlabel("Total Grade")
plt.ylabel("Number of students")
plt.yticks(np.arange(10))
plt.grid(axis='y')
```

```python
# Question 1 distribution
plt.figure(figsize=((10,5)))
n, bins, patches = plt.hist(x=grades.Q1, bins=10, color='blue',
                            alpha=0.7, rwidth=0.85)
plt.xlabel("Grade Question #1")
plt.ylabel("Number of students")
plt.yticks(np.arange(19))
plt.grid(axis='y')
```

```python
# Question 2 distribution
plt.figure(figsize=((10,5)))
n, bins, patches = plt.hist(x=grades.Q2, bins=10, color='g',
                            alpha=0.7, rwidth=0.85)
plt.xlabel("Grade Question #2")
plt.ylabel("Number of students")
plt.yticks(np.arange(13))
plt.grid(axis='y')
```

```python
# Question 3 distribution
plt.figure(figsize=((10,5)))
n, bins, patches = plt.hist(x=grades.Q3, bins=10, color='r',
                            alpha=0.7, rwidth=0.85)
plt.xlabel("Grade Question #3")
plt.ylabel("Number of students")
plt.yticks(np.arange(16))
plt.grid(axis='y')
```

```python
bins = np.arange(0,11)
plt.figure(figsize=((12,8)))
plt.hist([grades.Q1, grades.Q2, grades.Q3], bins=bins, color=['b','g','r'], label=["Q1", "Q2", "Q3"],
                            alpha=0.5, rwidth=0.85)

# plt.hist(x=grades.Q2, bins=bins, color='g', label="Q2",
#                             alpha=0.5, rwidth=0.85)

# plt.hist(x=grades.Q3, bins=bins, color='r', label="Q3",
#                             alpha=0.5, rwidth=0.85)
plt.legend()
plt.xticks(bins)
plt.yticks(np.arange(20))
plt.ylabel("Number of Students")
plt.xlabel("Grade")
plt.grid(axis='y')
```
